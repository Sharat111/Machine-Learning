There are some words like be, that, a, to and so on. These words are stop words. Generally, stop words should be removed to prevent them from affecting our results.

```python
from nltk.corpus import stopwords
 
stopwords.words('english')
```

```python
clean_words = text2[:]
 
for text in text2:
 
    if text in stopwords.words('english'):
 
        clean_words.remove(text)
```

### Sentiment Analysis using NaiveBayes - Example

```python
import nltk
from nltk.tokenize import word_tokenize
  
# Step 1 – Training data
train = [("the training went very well and i am glad it happened", "pos"),
  ("the number of participants were high hence i got no space to sit", "neg"),
  ("loved the examples, hated the slow speed", "neg"),
  ("the class was interesting but very noisy", "pos"),
  ("excellent class. will request some more notes for self reading", "pos"),
  ("the delivery of the lecture was easy to understand.", "pos"),
  ("we did not get many breaks, hence became exhausted quickly", "neg"),
]
  
# Step 2 
dictionary = set(word.lower() for comment in train for word in word_tokenize(comment[0]))
  
# Step 3
t = [({word: (word in word_tokenize(x[0])) for word in dictionary}, x[1]) for x in train]
  
# Step 4 – the classifier is trained with sample data
classifier = nltk.NaiveBayesClassifier.train(t)
  
test_data = "we did not get any time to practice."
test_data_features = {word.lower(): (word in word_tokenize(test_data.lower())) for word in dictionary}
  
print (classifier.classify(test_data_features))
```

### Synonyms & Antonyms using NLTK

```python
from nltk.corpus import wordnet
 
syn = wordnet.synsets("attack")
 
print(syn[0].definition())
 
print(syn[0].examples())
```

```python
syn = wordnet.synsets("NLP")
 
print(syn[0].definition())
 
syn = wordnet.synsets("Tiger")
 
print(syn[0].definition())
```

```python
synonyms = []
 
for syn in wordnet.synsets('world'):
 
    for lemma in syn.lemmas():
 
        synonyms.append(lemma.name())
 
print(synonyms)
```

```python
antonyms = []
 
for syn in wordnet.synsets("big"):
 
    for l in syn.lemmas():
 
        if l.antonyms():
 
            antonyms.append(l.antonyms()[0].name())
 
print(antonyms)
```

```python
# Difference between Lemmatization and Stemming

from nltk.stem import WordNetLemmatizer
 
from nltk.stem import PorterStemmer
 
stemmer = PorterStemmer()
 
lemmatizer = WordNetLemmatizer()
 
print(stemmer.stem('stones'))
 
print(stemmer.stem('speaking'))
 
print(stemmer.stem('bedroom'))
 
print(stemmer.stem('jokes'))
 
print(stemmer.stem('lisa'))
 
print(stemmer.stem('purple'))
 
print('----------------------')
 
print(lemmatizer.lemmatize('stones'))
 
print(lemmatizer.lemmatize('speaking'))
 
print(lemmatizer.lemmatize('bedroom'))
 
print(lemmatizer.lemmatize('jokes'))
 
print(lemmatizer.lemmatize('lisa'))
 
print(lemmatizer.lemmatize('purple'))
```
